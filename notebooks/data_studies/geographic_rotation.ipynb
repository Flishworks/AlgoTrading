```python
tickers = ['INDA', 'VNM', 'SPY', 'FXI', 'EWZ', 'VGK', 'EEM', 'VWO', 'EFA', 'IEFA', 'ILF', 'VXUS', 'EWJ', 'EWY', 'EIDO', 'THD', 'EPP', 'EWA', 'EWC', 'EWH', 'EWS', 'EWQ', 'EWL', 'EWN', 'EWO', 'EWP', 'EWD', 'EWI', 'EWL', 'EWU']
# get date 3 months ago 
start_date = datetime.today() - timedelta(days=120)
start_date = start_date.replace(hour=0, minute=0, second=0, microsecond=0)

# get data
ticker_data = {}
for ticker in tickers:
    ticker_data[ticker] = ldi.get_ticker(ticker, start_date=start_date, interval='1d', end_date = None)
```

### Plot trend over time

```python
i = 0
for ticker, data in ticker_data.items():
    plt.plot(data['close']/data['close'][0], label=ticker)
    i += 1
    if i % 10 == 0:
        #shoe legend on left side of the plot
        plt.legend(loc='upper left')
        plt.show()
plt.legend()
plt.show()
```
