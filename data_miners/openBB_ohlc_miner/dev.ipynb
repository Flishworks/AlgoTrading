```python
throttle = 5
try:
    while True:
        for symbol in config.symbols:
            try:
                price_data = obb.equity.price.quote(symbol="AAPL", provider="yfinance")
                dir = r'C:\Users\avido\Documents\other code\AlgoTrading\data'
                price_data.to_df().to_csv(os.path.join(dir, file), mode = "a+", header = not os.path.exists(file))
                print(symbol + " scraped at " + str(price_data.index[-1]))
                time.sleep(throttle) #for API throttling. Accounted for in sleep beliw
            except:
                print(f"Failed processing {symbol}")
        price_data
        time.sleep(10 - throttle*len(config.symbols))
except KeyboardInterrupt:
    print >> sys.stderr, '\nExiting by user request.\n'
    sys.exit(0)
```
